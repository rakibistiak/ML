```python
import numpy as np
data_1=np.random.normal(loc=0, scale=1, size=1000)
data_2=np.random.lognormal(mean=0, sigma=1, size=1000)
```

```python
from scipy.stats import skew
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df_1=pd.DataFrame(data_1, columns=["ages"])
df_2=pd.DataFrame(data_2, columns=["salary"])
df_2.head()
```

```python
skewness_1 = skew(df_1)
skewness_2 = skew(df_2)
mean_1=np.mean(df_1['ages'])
mean_2=np.mean(df_2['salary'])

median_1=np.median(df_1['ages'])
median_2=np.median(df_2['salary'])

mode_1=np.median(df_1['ages'])
mode_2=np.median(df_2['salary'])
```

```python
plt.figure(figsize=(7,4))
plt.hist(df_2, bins=20, alpha=0.6, color='blue' , label="Data")

plt.axvline(mean_2, color='green', linestyle='dashed', linewidth=2, label=(f'Mean: {mean_2:.2f}'))
plt.axvline(median_2, color='red', linestyle='dashed', linewidth=2, label=(f'Mean: {median_2:.2f}'))
plt.axvline(mode_2, color='yellow', linestyle='dashed', linewidth=5, label=(f'Mean: {mode_2:.2f}'))

plt.xlabel('Values')
plt.ylabel('Freequency')
plt.title("Right Skewed Histogram")
plt.legend(loc='upper right')
#sns.histplot(df_2)
plt.show()
```
